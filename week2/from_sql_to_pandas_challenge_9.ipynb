## 10. Select all store names, their store id and the discounts they offer.

````python
discounts.merge(stores,how='right',on='stor_id')[['stor_id','stor_name','discount']]
# display(discounts.head(2))
# stores.head(2)
````

### Hint:

```sql

RIGHT JOIN stores AS s
ON d.stor_id=s.stor_id;

```
