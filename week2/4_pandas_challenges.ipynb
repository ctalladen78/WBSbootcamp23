###### **Exercise 9:**
what brand has the worse CO2 Emissions on average?

```python
# your code here
cars.groupby('Make').agg({'CO2 Emission Grams/Mile':'mean'}).sort_values(by='CO2 Emission Grams/Mile',ascending=True).head(1)
```

###### **Exercise 10:**
did cars consume more gas in the eighties?

```python
# your code here
eighties = cars.groupby('Year').agg({'Fuel Barrels/Year':'sum'}).loc[1984:1989].sum()
not_eighties = cars.groupby('Year').agg({'Fuel Barrels/Year':'sum'}).loc[1989:2017].sum()
print(eighties)
print(not_eighties)
print('did cars consume more gas in the eighties?: No')
```

###### **Exercise 11:**
group cars by fuel type and aggregate them by the following criteria: 

- The maximum number cylinders
- The oldest year
- The average Miles Per Gallon in the city

```python
# your code here
print(cars.groupby('Fuel Type')['Cylinders'].max())
print(cars.groupby('Fuel Type')['Year'].max())
print(cars.groupby('Fuel Type')['City MPG'].mean())
```
