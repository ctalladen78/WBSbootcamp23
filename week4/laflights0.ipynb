```python
#
# 1. Carriers on time: given a destination and a departure date, 
# returns the three carriers with a lower departure delay in average.

# 2. Crowd avoidance: given a destination and a month, recommend 
# which are the 5 best days and time range (morning, afternoon, evening and night) 
# where there is a low number of flights departing from the airport, so there will be less people on there. 
# The output of this functions should be two functions called 
# carriers_on_time(dest,dep)
# crowd_avoidance(dest,mo)

# The time ranges will be:
# Morning: 5 am to 12 pm (noon)
# Afternoon: 12 pm to 5 pm
# Evening: 5 pm to 9 pm
# Night: 9 pm to 4 am
#
# Los Angeles International Airport (LAX)
# Ontario International Airport (ONT)
# John Wayne Airport (SNA)
# Hollywood Burbank Airport (BUR)
# Long Beach Airport (LGB)
```

```python
from datetime import datetime
# input (str), (datetime)
# returns the three carriers with a lower departure delay in average.
def carriers_on_time(dest,dept_str) : 
    # test input strings
    
    # match departure date & destination
    carriers_df = flights[flights['fl_datetime'].str.contains(dept_str)][['op_carrier','dep_delay','dest']][flights['dest']==dest].sort_values(by='dep_delay',ascending=True)
    # match carriers for dept date
    # df3 = carriers_df.merge(delays_df,on='op_carrier')
    # display(df3)
    # df = flights.groupby('op_carrier','dep_delay','dep_time','fl_datetime').agg({'dep_delay':'mean'})
    delays_df = carriers_df.groupby('op_carrier').agg({'dep_delay':'mean'}).sort_values(by='dep_delay',ascending=True)
    display(delays_df)



```

```python
# input (str), (int)
# return 5 (day, time range)
def crowd_avoidance(origin,mo) : 
    # test input strings
    # 'op_carrier','dep_delay','origin','fl_datetime'
    flights.loc[(flights['dep_time'] >500)&(flights['dep_time'] <=1200),'time_range'] = 'morning'
    flights.loc[(flights['dep_time'] >1200)&(flights['dep_time'] <=1800),'time_range'] = 'afternoon'
    flights.loc[(flights['dep_time'] >1800)&(flights['dep_time'] <=2400),'time_range'] = 'evening'
    flights.loc[(flights['dep_time'] >2400)&(flights['dep_time'] <=500),'time_range'] = 'night'
    # match month from date
    # departures_df = flights[(flights['dest']==origin)&(flights['fl_datetime'].str.contains('\d{4}-'+mo))]
    # departures_df = departures_df[['op_carrier','dest','fl_datetime','time_range']].value_counts().sort_values(ascending=True).head(5)
    departures_df = flights[(flights['origin']==origin)&(flights['fl_datetime'].str.contains('\d{4}-'+mo))]
    departures_df = departures_df[['origin','fl_datetime','time_range']].value_counts().sort_values(ascending=False).head(5)
    # filter from departures_df
    print(departures_df)
```
